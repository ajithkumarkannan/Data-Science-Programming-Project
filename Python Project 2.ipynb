```python
#loading file 4

import pandas as pd
import numpy as np

PatH = pd.read_csv('D:/MS Rutgers/Spring 2024/BINF 5900 Data Science Programming/Project/Final project/Patient health habits.csv')
PatH.drop_duplicates(inplace=True)
PatH
```

```python
#Merging patient with data file 4 to get the final dataframe 'patient_final'

import pandas as pd
import numpy as np

patient_final = pd.merge(patient,PatH,how='left',on='PID')

pd.set_option('display.max_columns', None)

patient_final
```

```python
# Calculate BMI and assign to 'BMI' column
patient_final['BMI'] = round(patient_final['Weight'] / (patient_final['Height'] ** 2), 2)

patient_final
```

```python
# Fill values for column Hypertensive based on the patient's vitals.

patient_final.loc[(patient_final['SBP'] >= 145) & (patient_final['DBP'] >= 80), 'Hypertensive'] = 'Yes'
patient_final.loc[(patient_final['SBP'] <= 145), 'Hypertensive'] = 'No'
patient_final
```

```python
#Replacing NaN values in column medication. We consider the patient is not taking any medication if the coloumn is left empty.

patient_final['Medication'].fillna('None', inplace=True)
patient_final
```

```python
# Define thresholds for each criterion to determine if the patient is at risk of developing cushing's disease
corticosteroid_threshold = 'CortSt'
hdl_threshold_male = 40  
hdl_threshold_female = 50  
total_cholesterol_threshold = 200  
triglyceride_threshold = 150 
glucose_threshold = 100  
sbp_threshold = 145
dbp_threshold = 80
waist_threshold_male = 40  
waist_threshold_female = 35  
bmi_threshold = 30 

# Assess risk based on criteria, considering sex and waist circumference
for i in range(len(patient_final)):
    row = patient_final.loc[i]
    if (row['Medication'] == corticosteroid_threshold and
        ((row['Gender'] == 'Male' and row['HDL levels'] < hdl_threshold_male) or
         (row['Gender'] == 'Female' and row['HDL levels'] < hdl_threshold_female)) and
        row['Total Cholesterol levels'] > total_cholesterol_threshold and
        row['TG levels'] > triglyceride_threshold and
        row['FBG'] > glucose_threshold and
        row['SBP'] > sbp_threshold and
        row['DBP'] > dbp_threshold and
        ((row['Gender'] == 'Male' and row['Waist Circumference'] > waist_threshold_male) or
         (row['Gender'] == 'Female' and row['Waist Circumference'] > waist_threshold_female)) and
        row['BMI'] > bmi_threshold):
        patient_final.loc[i, 'Cushing risk'] = 'Yes'
    else:
        patient_final.loc[i, 'Cushing risk'] = 'No'

patient_final
```

```python
# Convert 'Waist Circumference' from inches to centimeters and round off to two decimal points
patient_final['Waist Circumference'] = round(patient_final['Waist Circumference'] * 2.54, 2)  # 1 inch = 2.54 cm

# Convert 'Height' from meters to centimeters and round off to two decimal points
patient_final['Height'] = round(patient_final['Height'] * 100, 2)  # 1 meter = 100 cm


# Iterate over rows and calculate siMS score
for i in range(len(patient_final)):
    row = patient_final.loc[i]
    waist_cm = row['Waist Circumference']  # in cms
    height_cm = row['Height']  # in cms
    fbg = row['FBG']  # Fasting blood glucose
    tg = row['TG levels']  # Triglyceride levels
    sbp = row['SBP']  # Systolic blood pressure
    hdl_threshold = 40 if row['Gender'] == 'Male' else 50
    hdl = row['HDL levels']  # HDL cholesterol levels

    # Calculate siMS score
    siMS_score = round(2 * (waist_cm / height_cm) + fbg / 100 + tg / 150 + sbp / 130 + hdl / hdl_threshold, 2)

    # Assign siMS score to the corresponding row
    patient_final.loc[i, 'siMS score'] = siMS_score

patient_final
```

```python
#Two sample T-test using BMI and Gender

from scipy.stats import ttest_ind

Men = patient_final[patient_final['Gender'] == 'Male']['BMI']
Women = patient_final[patient_final['Gender'] == 'Female']['BMI']

Mean = ttest_ind(Men,Women)

print(Mean)
```

```python
#Two sample T-test using Gender and siMS score

from scipy.stats import ttest_ind

Men = patient_final[patient_final['Gender'] == 'Male']['siMS score']
Women = patient_final[patient_final['Gender'] == 'Female']['siMS score']

Mean_siMS = ttest_ind(Men,Women)

print(Mean_siMS)
```

```python
#One way ANOVA using BMI and PhysAct levels

from scipy.stats import f_oneway

PA_low = patient_final[patient_final['PhysAct'] == 'Low']['BMI']
PA_mod = patient_final[patient_final['PhysAct'] == 'Moderate']['BMI']
PA_high = patient_final[patient_final['PhysAct'] == 'High']['BMI']

result = f_oneway(PA_low, PA_mod, PA_high)

print(result)
```

```python
#Two way ANOVA to see possible effect of PhysAct, Medication, and PhysAct + Medication on siMS score

import statsmodels.api as sl
from statsmodels.formula.api import ols

formula = 'Q("siMS score") ~ C(PhysAct) + C(Medication) + C(PhysAct):C(Medication)'

model = ols(formula, data=patient_final).fit()

anova_table = sl.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
#Linear regression analysis

import statsmodels.api as ss

X = patient_final[['Waist Circumference', 'SBP', 'HDL levels']]  # Independent variables
y = patient_final['siMS score']  # Dependent variable

X = ss.add_constant(X)
model = ss.OLS(y, X).fit()

model.summary()
```
